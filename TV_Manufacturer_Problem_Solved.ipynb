```
import numpy as np
import sympy as sym

x1, x2 = sym.symbols('x1, y2')
f = (339 - .01*x1 - .003*x2)*x1 + (399 - .01*x2 - .004*x1)*x2 - (400000 + 195*x1 + 225*x2)
from sympy.plotting import plot3d  
plot3d(f,(x1,0,10000),(x2,0,10000));

```

```
#use Matplotlib to show the contour plot

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10000, 10) #creates an array from 0 to 5000 (spacing = 1)
y = np.arange(0, 10000, 10)
xm, ym = np.meshgrid(x, y)
#zm = xm**2 +  ym**2
zm = (339 - .01*xm - .003*ym)*xm + (399 - .01*ym - .004*xm)*ym - (400000 + 195*xm + 225*ym)
#zm = xm**4 +  ym**4 - 4*xm*ym + 1 #hard-code function
cf = plt.contourf(xm,ym,zm,20)
cb = plt.colorbar(cf)
cb.set_label(label = 'Profit ($)', weight = 'bold', size = 16)
plt.axis('equal')
plt.xlabel('x1 (# of 19-inch TV sets)',fontsize = 16)
plt.ylabel('x2 (# of 21-inch TV sets)',fontsize = 16)
plt.show()
```
